### Section 2: Exploratory Data Analysis

In this section, our main goal is to understand the structure of our data better and gain some insights of the vehicle insurance fraud dataset by drawing visuals and representing columns and their relationship.

Let's look at each numeric column individually and draw their histogram: 

```python
df_numeric = df.select_dtypes(include="number")

# for each numeric column, create a subplot
fig, axes = plt.subplots(4,3, figsize=(20,20))

# Creating the for loop
for plot_num, col in enumerate(df_numeric.columns, start=1):
    
    # draw histogram of the current column
    plt.subplot(4,3,plot_num)
    plt.hist(df[col], color='lightblue', bins=24)
    
    # Set plot attributes:
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f"Distribution of {col}")
    
    # Draw vertical lines for mean and median of numeric cols
    plt.axvline(df[col].mean(), c='darkblue', label='Mean')
    plt.axvline(df[col].median(), c='green', label='Median')
    plt.legend()
    
# tighten subplots    
plt.tight_layout()


# hiding the empty plots:
fig.delaxes(axes[3][2])
fig.delaxes(axes[3][1])
plt.show()    
```

We can see from the plots that: 

- The dataset records a history of policies over a span of three years in the previous century (1994 to 1996) with more than 6000 records for 1994 and around 5000 for the year 1995 reaching to almost 4000 records for the following year, 1996.
- The distribution of 'WeekOfMonth' and "WeekOfMonthClaimed variables indicates that each of the first four weeks has more than 3000 records. Only for the fifth month of the week (which is not very common to happen), do we have at most 1500 (almost half) records. 
- The distribution of our target variable, i.e., 'FraudFound_P' variable, shows the dataset is not equally balanced for the valid and invalid insurance claims. The majority ofthe dataset, more than 14k rows, indicate there were zero probability for a fraud to have happened, except for almost 1500 rows, which are suspicious of being fraudulent insurance claims.
- The variable 'Age' is displayed in an approximately normal distribution.
- The variable 'DriverRating' is displaying an almost uniform distibution for every unique rating associated with the driver's quality of driving. 
- The variable 'Deductible' is highly focused on the value 400, having the majority of rows, around 14k rows of the 15k-row dataset. The other deductible values are very unlikely to be observed, way more less than 2k rows of the data combined. 
- The variable 'RepNumber' also shows somewhat of a uniform distribution. Every unique RepNumber in the dataset appear in around 10000 rows of the rows in our dataset.

```python
import plotly.express as px

# Calculate the percentages
no_fraud_percentage = len(df[df['FraudFound_P'] == 0]) / len(df) * 100
fraud_percentage = len(df[df['FraudFound_P'] == 1]) / len(df) * 100

# Create a new DataFrame for plotting
cats = {
    'Category': ['No Fraud', 'Fraud'],
    'Percentage': [no_fraud_percentage, fraud_percentage]
    }

print(no_fraud_percentage, '% of records not commited fraud out of all')
print(fraud_percentage, '% of records did commit a fraud out of all')

plot_df = pd.DataFrame(cats)

# Create a bar plot using Plotly Express
fig = px.bar(plot_df, x='Category', y='Percentage', text='Percentage',
             labels={'Percentage': 'Percentage of Records (%)'},
             title='Fraud Distribution')

# Show the plot
fig.show()
```

Let's calculate the p values: 

```python
from scipy import stats

df_numeric = df_clean.select_dtypes(include="number")
target=['FraudFound_P','Fault']
for col in target:
    print(f'\n Correlation between {col} and \n')
    for column in df_numeric.columns:
        if column in ['FraudFound_P','Fault']:
            continue
        else:
            statistic,p = stats.pearsonr(df_clean[column],df_clean[col])
            print(f"\t {column}: statistic={statistic:4.3f}, p_value={p:4.3f}")
```

```python
# Create a dictionary to store the T-statistic values
statistics = {}
df_numeric = df_clean.select_dtypes(include="number")
# Calculate T-statistic for each variable
for i in target:
    statistics[i] = {}
    for column in df_numeric.columns:
        if column in ['FraudFound_P', 'Fault']:
            continue
        else:
            statistic, p = stats.pearsonr(df_clean[column], df_clean[i])
            statistics[i][column] = statistic

# Create a subplot with three bar plots
fig, axs = plt.subplots(2, 1, figsize=(9, 15))

# Plotting bar plots for each target variable
for idx, i in enumerate(target):
    axs[idx].bar(statistics[i].keys(), statistics[i].values())
    axs[idx].set_xlabel('Variables')
    axs[idx].set_ylabel('statistic')
    axs[idx].set_title(f'statistic ( {i} vs predictors)')
    axs[idx].tick_params(axis='x', rotation=90) 
# Adjust layout and spacing
plt.tight_layout()

# Show the plot
plt.show()
```
