```python
for i in features:
    print(i)
    print(round(df[i].value_counts()/len(df)*100,2))
    print("="*50)
```

After exploring we decide to remove those category that are less than 2% in our data for each feature. 
If after removing, there exists only one category (like for the AddressChange_Claim column) we will drop that feature only entirely from the dataset, becuae it now can hold only one value among all possible category values. 

```python
print(df.shape)
df_clean = df.copy(deep=True)
for col in features:
    percentage = round(df[col].value_counts()/len(df)*100,2)
    remove = percentage[percentage <= 2].index
    # if the current feature value is among those categories to be removed, 
    # then exclude (by using tilda ~) them from the dataset 
    df_clean = df_clean[~df_clean[col].isin(remove)]
    
for col in features:
    # count the number of unique remaining categories     
    unique = df_clean[col].nunique()
    # if the number of unique categories is either one or zero, then exclude that column from the dataset     
    if unique <= 1:
        df_clean.drop(columns=[col], inplace=True)
        
# df_clean.columns
df_clean.shape
```

```python
features = [i  for i in df_clean.columns if df_clean[i].dtype==object and i != 'FraudFound_P']

# Calculate the number of rows and columns for the grid
num_rows = 8
num_cols = 3

# Create subplots with the specified layout
fig, axes = plt.subplots(num_rows, num_cols, figsize=(10, 30))
plt.subplots_adjust(wspace=0.4, hspace=0.5)

# Loop through features and create countplots in the subplots
for i, feature in enumerate(features):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]
    
    sns.countplot(x=feature, data=df_clean, ax=ax,hue='FraudFound_P', palette=["lightblue","#ffb266"])
    ax.set_title(f'Countplot of {feature}')
    ax.set_xlabel(f'{feature}')
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=45)
    
# Hide any empty subplots
for i in range(len(features), num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    fig.delaxes(axes[row, col])

plt.tight_layout()
plt.show()
```

Now let's do some statistical analysis and preprocessing of data to convert categorical columns to numeric ones te prepare everything for the next step, model training.

### Section 3: Statistical Analysis and Preprocessing Data for Modeling

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder

def preprocess_dataset(df):
    # find all categorical columns having only two categories      
    two_categories = [i for i in df_clean.columns if df_clean[i].nunique() == 2]

    # Initialize a LabelEncoder instance from scikit-learn library 
    labelEncoder = LabelEncoder()

    # Iterate through each two_category column
    for column in two_categories:
        # Fit the label encoder and transform the column values
        df_clean[column] = labelEncoder.fit_transform(df_clean[column]) 
    
    '''
    The calculated frequency encoding values are then mapped 
    to the respective columns in the DataFrame. This technique can be used to encode categorical 
    features based on the relative frequencies of their categories in different target classes.

    '''
    
    # List of columns for frequency encoding
    months_weeks = ['WeekOfMonth', 'Month', 'DayOfWeek', 'DayOfWeekClaimed', 'MonthClaimed', 'WeekOfMonthClaimed']

    # Apply frequency encoding for Day, Month and Week Categorical Columns based on the target variable (FraudFound_P): 
    for column in months_weeks:
        # Calculate the frequency of each category for FraudFound_P=1 and FraudFound_P=0
        freq_1 = df_clean[df_clean['FraudFound_P'] == 1].groupby(column).size() / len(df)
        freq_0 = df_clean[df_clean['FraudFound_P'] == 0].groupby(column).size() / len(df)

        # Calculate the frequency encoding value using the formula (15*freq_1 + freq_0) / 2
        freq_encoding = (15 * freq_1 + freq_0) / 2

        # Map frequency encoding to corresponding columns
        df_clean[column] = df_clean[column].map(freq_encoding) 
    
    #Encoding AgeOfPolicyHolder
    # age_mapping = {
    #     '31 to 35': 0,
    #     '36 to 40': 1,
    #     '41 to 50': 2,
    #     '51 to 65': 3
    # }
    # df_clean['AgeOfPolicyHolder'] = df_clean['AgeOfPolicyHolder'].map(age_mapping)

    #Encoding VehiclePrice           
    encoding_mp = {
        'less than 20000': '0',
        '20000 to 29000': '1',
        '30000 to 39000': '2', 
        '40000 to 59000': '3', 
        'more than 69000': '4'
    }
    df_clean['Vehicle_Price'] = df_clean['VehiclePrice'].map(encoding_mp)
    df_clean['Vehicle_Price'] = pd.to_numeric(df_clean['Vehicle_Price'])
    df_clean.drop('VehiclePrice',axis=1,inplace=True)

    #Encoding AgeOfVehicle
    age_mapping = {  
        'new': '0', 
        '5 years': '1',
        '6 years': '2',
        '7 years': '3',
        'more than 7': '4'
    }
    df_clean['AgeVehicle'] = df_clean['AgeOfVehicle'].map(age_mapping)
    df_clean['AgeVehicle'] = pd.to_numeric(df_clean['AgeVehicle'])
    df_clean.drop('AgeOfVehicle',axis=1,inplace=True)

    #Encoding NumberOfSuppliments
    suppliments_mapping = {
        'none': '0',
        '1 to 2': '1',
        '3 to 5': '2', 
        'more than 5': '3'
    }
    df_clean['Suppliments'] = df_clean['NumberOfSuppliments'].map(suppliments_mapping)
    df_clean['Suppliments'] = pd.to_numeric(df_clean['Suppliments'])
    df_clean.drop('NumberOfSuppliments',axis=1,inplace=True)

    # Encoding PastNumberOfClaims
    claim_mapping = {
        'none': '0',
        '1': '1',
        '2 to 4': '2',
        'more than 4': '3'
    }
    df_clean['PastClaims'] = df_clean['PastNumberOfClaims'].map(claim_mapping)
    df_clean['PastClaims'] = pd.to_numeric(df_clean['PastClaims'])
    df_clean.drop('PastNumberOfClaims',axis=1,inplace=True)
    
    
preprocess_dataset(df_clean)
```
